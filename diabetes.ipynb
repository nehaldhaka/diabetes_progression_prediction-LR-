```python
#importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#loading the dataset 
data=pd.read_csv("diabetes_linear_regression_dataset.csv")
data.head()
```

```python
#taking the input and output 
x=data[["insulin"]]
y=data["disease_progression"]
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(data['insulin'],data['disease_progression'])

m,c=np.polyfit(data['insulin'],data['disease_progression'],1)

plt.plot(x,m*data['insulin']+c)

# m is slope
```

```python
#visualisation using seaborn

sns.regplot(x="insulin",y="disease_progression",color="pink",data=data)
```

```python
#training and split the data 

from sklearn.model_selection import train_test_split

x=data["insulin"]
y=data["disease_progression"]
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
x_train , x_test , y_train , y_test
```

```python
print("---------x_train----------")
print(x_train)
print("----------y_train---------")
print(y_train)

print("----------x_test---------")
print(x_test)
print("----------y_test---------")
print(y_test)
```

```python
#chossing a model 
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train.values.reshape(-1,1),y_train)
model.predict([[150]])
```

```python
y_pred=model.predict(x_test.values.reshape(-1,1))
y_pred
```

```python
plt.scatter(x_test,y_test,color='pink')
plt.plot(x_test,y_pred,color='blue')
```
